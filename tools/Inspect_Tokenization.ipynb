```python
from megatron.tokenizer import build_tokenizer
from preprocess_data_loss_mask import Encoder
import jsonlines

from tokenizers import Tokenizer
from transformers import AutoTokenizer
```

```python
def print_lists(list1, list2):
    """
    This function takes two lists of the same length and prints the individual items exactly under each other.
    """
    if len(list1) != len(list2):
        raise ValueError("The two lists must be of the same length.")
    else:
        for i in range(len(list1)):
            print(f"{list1[i]}\t{list2[i]}")
```

```python
class AttrDict(dict):
    def __init__(self, *args, **kwargs):
        super(AttrDict, self).__init__(*args, **kwargs)
        self.__dict__ = self
```
